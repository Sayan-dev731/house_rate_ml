```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('home_dataset.csv')
```

```python
house_price = data['HousePrice'].values
house_size = data['HouseSize'].values
```

```python
# np.random.seed(0)
# house_size = np.random.randint(800, 5000, 100)
```

```python
plt.scatter(house_size, house_price, marker = '*', color = 'red')
plt.title('House Price vs House Size')
plt.xlabel('House Size (Sq.ft)')
plt.ylabel('house Price ($)')
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(house_size, house_price, test_size = 0.2, random_state = 42)
```

```python
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

```python
print("x_test shape:", x_test.shape)
print("y_test shape:", y_test.shape)
print("predictions shape:", prediction.shape)
```

```python
plt.scatter(x_test, y_test, marker = '*', color = 'green', label = 'Actual Price') 
plt.plot(x_test, prediction, color = 'red', linewidth = 1.5, label = 'Predicted Price')
plt.title('House Price Prediction')
plt.xlabel('House Size (Sq.ft)')
plt.ylabel('House Price (millions $)')
plt.legend()
plt.show()
```
